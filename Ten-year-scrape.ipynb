```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
years = [
    '2009',
    '2011',
    '2013',
    '2015',
    '2017',
    '2019',
    '2021'
]
```

```python
def number_of_results_in_year(year):
    year_href = "https://www.nysenate.gov/search/legislation?sort=desc&type=f_bill&searched=true&bill_session_year=" + str(year) + "&bill_status=SIGNED_BY_GOV&page=1"
    raw_html = requests.get(year_href).content
    soup_year = BeautifulSoup(raw_html, "html.parser")
    
    results_count_element = soup_year.find("div", class_="ds-search-extra")
    results_count_string = results_count_element.text.strip()
    results_count = re.findall("[0-9]+", results_count_string)
    results_count_number = int(results_count[0])
    total_pages = int(results_count_number / 10) + 1
    return total_pages
```

```python
def signed_bills(href):
    raw_html = requests.get(href).content
    soup_doc = BeautifulSoup(raw_html, "html.parser")

    all_bills = soup_doc.find_all(
        "div", class_="c-block c-list-item c-block-legislation"
    )

    signed_bills_list_of_dicts = []

    for bill in all_bills:
        bill_number = (
            bill.find("h3", class_="c-bill-num")
            .text.replace("\n", "")
            .replace("Bill ", "")
            .strip()
        )
        bill_link = bill.find("h3", class_="c-bill-num").find("a")["href"]
        bill_description = bill.find("p", class_="c-bill-descript").text.replace(
            "\n", ""
        ).strip()
        bill_date = (
            bill.find("p", class_="c-bill-update--date")
            .text.replace("\n", "")
            .replace("|  Signed by Governor", "")
        )
        bill_sponsor_tag = bill.find("p", class_="c-bill-update--sponsor")
        if bill_sponsor_tag != None:
            bill_sponsor = bill_sponsor_tag.text.replace("\n", "").replace(
                "Sponsor:", ""
            )
        else:
            bill_sponsor = ""

        bill_complete_link = f"https://www.nysenate.gov{bill_link}"
        bill_section_html = requests.get(bill_complete_link).content
        bill_section_soup = BeautifulSoup(bill_section_html, "html.parser")
        bill_section_element = bill_section_soup.find_all("div", class_="c-block c-bill-section c-bill--details")
        if len(bill_section_element) == 0:
            print("Not found")
        else:
            section_text = bill_section_element[0].find_all("dd")
            if len(section_text) == 0:
                print("Error :(")
            else:
                if len(section_text) == 1:
                    bill_section = ""
                else:
                    bill_section = section_text[1].text

        
        signed_bills_list_of_dicts.append(
            {
                "bill_number": bill_number,
                "bill_link": "https://www.nysenate.gov" + bill_link,
                "bill_description": bill_description,
                "bill_date": bill_date,
                "bill_section": bill_section,
                "bill_sponsor": bill_sponsor,
            }
        )

    return signed_bills_list_of_dicts
```

```python
page_numbers_main_list = []

for year in years:
    pages_to_loop = number_of_results_in_year(year)
    page_numbers = range(1, pages_to_loop + 1)
    for page in page_numbers:
        href = f"https://www.nysenate.gov/search/legislation?sort=desc&type=f_bill&searched=true&bill_session_year={year}&bill_status=SIGNED_BY_GOV&page=" + str(page)
        bill_list = signed_bills(href)
        page_numbers_main_list.extend(bill_list)

        print(f"calling signed bills for year {year} in page {page}")
```

```python
import pandas as pd
df = pd.DataFrame(page_numbers_main_list)
df
```
